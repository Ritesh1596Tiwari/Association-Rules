```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data =pd.read_csv("book.csv")
data
```

```python
data.info()
```

```python
#APRIORI ALGORITHM
```

```python
# 5% support

frequent_itemsets = apriori(data,min_support=0.05,use_colnames=True)
frequent_itemsets
```

```python
#With 80% confidence
rules_80=association_rules(frequent_itemsets,metric='lift',min_threshold=0.8)
rules_80
```

```python
plt.scatter(rules_80['support'],rules_80['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
# 10% support

frequent_itemsets = apriori(data,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# 70% confidence

rules = association_rules(frequent_itemsets,metric='lift', min_threshold=0.7)
rules
```

```python
#LEVERAGE VALUE 0 INDICATES INDEPENDENCE AND HIGH CONVICTION VALUE MEANS CONSEQUENT IS HIGHLY DEPENDEDING ON ANTECEDENT
```

```python
rules.sort_values('lift',ascending=False)
```

```python
rules[rules.lift>1]
```

```python
#VISUALIZATION
```

```python
plt.scatter(rules['support'], rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
